```python
import pandas as pd

data = pd.read_csv('drive/My Drive/nuclear_explosions.csv')
```

```python
print(data.columns)
```

```python
missing_values=data.isna().sum()
```

```python
missing_values
```

```python
duplicates = data[data.duplicated()]
print("Duplicate Rows:\n", duplicates)
```

```python
unique_values = data['Data.Purpose'].unique()
print("Unique Values in 'Data.Purpose':\n", unique_values)
```

```python
data_types = data.dtypes
print("Data Types of Columns:\n", data_types)
```

```python
import matplotlib.pyplot as plt
country_count = data['WEAPON SOURCE COUNTRY'].value_counts()
country_count.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Country-wise Count of Nuclear Explosions')
plt.show()
```

```python
top_n = 10  # Adjust as needed
location_count = data['WEAPON DEPLOYMENT LOCATION'].value_counts().head(top_n)
location_count.plot(kind='bar')
plt.xlabel('Deployment Location')
plt.ylabel('Count')
plt.title('Top {} Deployment Locations Count'.format(top_n))
plt.show()
```

```python
year_count = data['Date.Year'].value_counts().sort_index()
year_count.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Year-wise Count of Nuclear Explosions')
plt.show()
```

```python
country_avg_yield = data.groupby('WEAPON SOURCE COUNTRY')['Data.Yeild.Lower'].mean()
country_avg_yield.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Average Yield')
plt.title('Country-wise Average Explosion Yield')
plt.show()
```

```python
surface_count = data['Data.Type'].value_counts()
surface_count.plot(kind='bar')
plt.xlabel('Surface Type')
plt.ylabel('Count')
plt.title('Surface Type Count')
plt.show()
```

```python
surface_avg_yield = data.groupby('Data.Type')['Data.Yeild.Lower'].mean()
surface_avg_yield.plot(kind='bar')
plt.xlabel('Surface Type')
plt.ylabel('Average Yield')
plt.title('Surface Type-wise Average Explosion Yield')
plt.show()
```

```python
data_purpose_count = data['Data.Purpose'].value_counts()
data_purpose_count.plot(kind='bar')
plt.xlabel('Data Purpose')
plt.ylabel('Count')
plt.title('Data Purpose Distribution')
plt.show()
```

```python
import seaborn as sns
correlation = data[['Data.Yeild.Lower', 'Location.Cordinates.Depth']].corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Yield and Depth')
plt.show()
```

```python
data_numeric = data.select_dtypes(include=['float64', 'int64'])
data_numeric.boxplot()
plt.title('Box Plot of Numeric Variables')
plt.show()
```
